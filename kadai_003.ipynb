```python
import numpy as np
import pandas as pd
import openpyxl
```

```python
df1 = pd.read_excel("2022_年間売上表.xlsx")
df1
```

```python
df2 = pd.read_excel("2023_年間売上表.xlsx")
df2
```

```python
df3 = pd.concat([df1,df2],axis=0,ignore_index=True)
df3
```

```python
grouped = df3.groupby(["商品","売上年"]).sum()
grouped
```

```python
df = grouped_df3 = grouped.reset_index()
print(df)
```

```python
df = pd.DataFrame(data = grouped_df3)
```

```python
df.to_excel("売上一覧表.xlsx",sheet_name="Sheet",index=False)
```

```python
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

wb = load_workbook("売上一覧表.xlsx")
ws = wb.active

cell_range = ws["A1:C1"]
pattern_fill = PatternFill(patternType="solid" ,fgColor="F2F2F2")
for row in cell_range:
    for cell in row:
        cell.fill = pattern_fill

wb.save("売上一覧表1.xlsx")
```
